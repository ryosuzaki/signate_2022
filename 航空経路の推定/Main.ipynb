```python
!pip install xmltodict
import urllib
import xmltodict
import json

def only_kana(word):
    alpha=list("abcdefghijklmnopqrstuvwxyz")
    for s in word:
        if s in alpha:
            return False
    return True

def get_suggestion(word):
    response = urllib.request.urlopen('https://www.google.com/complete/search?hl=en&output=toolbar&q='+urllib.parse.quote(word))
    content = response.read()
    response.close()
    dict_xml=xmltodict.parse(content)
    
    if dict_xml['toplevel'] is None:
        return None
    else:
        if len(dict_xml['toplevel']['CompleteSuggestion'])>1:
            return dict_xml['toplevel']['CompleteSuggestion'][0]['suggestion']['@data']
        else:
            return dict_xml['toplevel']['CompleteSuggestion']['suggestion']['@data']
```

```python
#https://dev.classmethod.jp/articles/pytrends_introduction/#toc-2
from pytrends.request import TrendReq
pytrend = TrendReq(hl='ja-jp',tz=540)

kw_idx=[]
kw_list=[]
for i in range(len(test_df)):
    if test_df.loc[i,"suggested_word"] is not None:
        kw_idx.append(i)
        kw_list.append(test_df.loc[i,"suggested_word"])
    else:
        test_df.loc[i,"N_10"]=N_mean
        test_df.loc[i,"E_10"]=E_mean
kw_list=[kw_list[idx:idx+5] for idx in range(0,len(kw_list),5)]
trend_dfs=[0]*len(kw_list)
for i in range(len(kw_list)):
    # googleにリクエストする。
    pytrend.build_payload(kw_list=kw_list[i], timeframe='2018-07-01 2022-07-01', geo="JP")
    # 小区域別のインタレストの取得
    trend_dfs[i]=pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=True)
```

```python
trends=[]
for i in range(len(trend_dfs)):
    for j in range(len(trend_dfs[i].columns)-1):
        trends.append(trend_dfs[i].iloc[:,1+j])
```

```python
city_names=[]
for tmp_df in trends:
    city_names.append(tmp_df.loc[tmp_df>50].index.values)
```

```python
import numpy as np
loc=[]
for names,i in zip(city_names,kw_idx):
    if len(names)!=0:
        tmp_loc=[]
        for name in names:
            tmp_loc.append([loc_df.loc[name,"lat"],loc_df.loc[name,"long"]])
        tmp_loc=np.array(tmp_loc)
        lat,lon=np.mean(tmp_loc, axis=0)
    else:
        lat,lon=N_mean,E_mean
    test_df.loc[i,"N_10"]=lat
    test_df.loc[i,"E_10"]=lon

test_df        
```
